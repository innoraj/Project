**Selecting Best Features**

```
#Feature selection
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

data = pd.read_csv('https://raw.githubusercontent.com/innoraj/Project/master/Dataset1.csv')
X = data.iloc[:,0:13]  #independent columns
y = data.iloc[:,-1]    #target column 
#apply SelectKBest class to extract top 10 best features
bestfeatures = SelectKBest(score_func=chi2, k=10)
fit = bestfeatures.fit(X,y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  #naming the dataframe columns
print(featureScores.nlargest(10,'Score'))  #print 10 best features
```

**Classification Algorithm**

```
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import SGDClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
#from sklearn.svm import SVC

data = pd.read_csv('https://raw.githubusercontent.com/innoraj/Project/master/Dataset1.csv')

X = data.iloc[:,[1,6,5,12,0,4,7,8,11,9]]
y = data.iloc[:,12]

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.8)

modelLR = LogisticRegression()
modelLR.fit(X_train, y_train)

modelNB = GaussianNB()
modelNB.fit(X_train, y_train)

modelSGD = SGDClassifier(loss='hinge', shuffle=True, random_state=101)
modelSGD.fit(X_train, y_train)

modelKN = KNeighborsClassifier(n_neighbors=15)
modelKN.fit(X_train, y_train)

modelDT = DecisionTreeClassifier(max_depth=10, random_state=101,max_features = None, min_samples_leaf = 15)
modelDT.fit(X_train, y_train)

modelRF = RandomForestClassifier(n_estimators=70, oob_score=True, n_jobs=-1,random_state=101, max_features = None, min_samples_leaf = 30)
modelRF.fit(X_train, y_train)

#modelSV = SVC(kernel="linear", C=0.025,random_state=101)
#modelSV.fit(X_train, y_train)


print(" LogisticRegression ", modelLR.score(X_test, y_test)*100)
LogisticRegression = modelLR.score(X_test, y_test)*100
print(" GaussianNB ", modelNB.score(X_test, y_test)*100)
GaussianNB = modelNB.score(X_test, y_test)*100
print(" SGDClassifier ", modelSGD.score(X_test, y_test)*100)
SGDClassifier = modelSGD.score(X_test, y_test)*100
print(" KNeighborsClassifier ", modelKN.score(X_test, y_test)*100)
KNeighborsClassifier = modelKN.score(X_test, y_test)*100
print(" DecisionTreeClassifier ", modelDT.score(X_test, y_test)*100)
DecisionTreeClassifier = modelDT.score(X_test, y_test)*100
print(" RandomForestClassifier ", modelRF.score(X_test, y_test)*100)
RandomForestClassifier = modelRF.score(X_test, y_test)*100
#print(" SupportVectorClassifier ", modelSV.score(X_test, y_test)*100)
#SupportVectorClassifier = modelSV.score(X_test, y_test)*100
```

**Confusion Matrix for Logistic Regression**

```
import time
start = time.time()
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
data = pd.read_csv('https://raw.githubusercontent.com/innoraj/Project/master/Dataset1.csv')
X = data.drop(['cardio'], axis=1)
y = data.cardio.copy()
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.8)
model = LogisticRegression()
model.fit(X_train, y_train)
y_pred=model.predict(X_test)
end = time.time()
print("Execution time in Seconds ", end-start)
print("Accuracy ", model.score(X_test, y_test)*100)
#print(data.shape)
```

```
# import the metrics class
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```
# import required modules
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```
class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```
print("Accuracy:",metrics.accuracy_score(y_test, y_pred)*100)
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
```

```
y_pred_proba = model.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

**Ensemble Technique**

```
import time
start = time.time()
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import VotingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import tree
data = pd.read_csv('https://raw.githubusercontent.com/innoraj/Project/master/Dataset1.csv')

X = data.drop(['cardio'], axis=1)
y = data.cardio.copy()

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.8)

model1 = LogisticRegression(random_state=1)
model2 = tree.DecisionTreeClassifier(random_state=1)
model3 = GaussianNB()
model = VotingClassifier(estimators=[('lr', model1), ('dt', model2),('nb', model3)], voting='hard')
model.fit(X_train,y_train)
model.score(X_test,y_test)

end = time.time()
print("Execution time in Seconds ", end-start)
print("Accuracy ", model.score(X_test, y_test)*100)
```
