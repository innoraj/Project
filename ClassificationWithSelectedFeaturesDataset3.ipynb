```
import pandas as pd
import numpy as np
np.random.seed(42)

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

data = pd.read_csv('https://raw.githubusercontent.com/innoraj/Project/master/Dataset3.csv')

X = data.iloc[:,[7,9,11,2,8,4,0,3,10,1]]
y = data.iloc[:,13]

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.8)

modelLR = LogisticRegression()
modelLR.fit(X_train, y_train)

modelNB = GaussianNB()
modelNB.fit(X_train, y_train)

modelKN = KNeighborsClassifier(n_neighbors=5)
modelKN.fit(X_train, y_train)

modelDT = DecisionTreeClassifier(max_depth=10, random_state=42,max_features = None, min_samples_leaf = 15)
modelDT.fit(X_train, y_train)

modelSV = SVC(kernel="linear", C=0.025, random_state=42)
modelSV.fit(X_train, y_train)


print(" LogisticRegression ", modelLR.score(X_test, y_test)*100)
LogisticRegression = modelLR.score(X_test, y_test)*100
print(" GaussianNB ", modelNB.score(X_test, y_test)*100)
GaussianNB = modelNB.score(X_test, y_test)*100
print(" KNeighborsClassifier ", modelKN.score(X_test, y_test)*100)
KNeighborsClassifier = modelKN.score(X_test, y_test)*100
print(" DecisionTreeClassifier ", modelDT.score(X_test, y_test)*100)
DecisionTreeClassifier = modelDT.score(X_test, y_test)*100
print(" SupportVectorClassifier ", modelSV.score(X_test, y_test)*100)
SupportVectorClassifier = modelSV.score(X_test, y_test)*100
```

```
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
objects = ('Logistic Regression','GaussianNB','KNeighbors Classifier','DecisionTree Classifier', 'Support Vector Classifier')
y_pos = np.arange(len(objects))
performance = [LogisticRegression,GaussianNB,KNeighborsClassifier,DecisionTreeClassifier, SupportVectorClassifier]
plt.bar(y_pos, performance, align='center', alpha=0.5,color='blue') 
plt.xticks(y_pos, objects,rotation=45)
plt.ylabel('Accuracy')
plt.title('Comparison of Model for Heart Attack Prediction Dataset 3')
plt.show()
plt.savefig("Comparison of Model for Heart Attack Prediction Dataset 3.png")
files.download("Comparison of Model for Heart Attack Prediction Dataset 3.png")
```
