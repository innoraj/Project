```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Loading the Data Set
data = pd.read_csv('https://raw.githubusercontent.com/innoraj/Project/master/Dataset1.csv')
```

```python
# To See the dimension of the dataset
print(data.shape)
```

```python
# See the column names
print(data.columns)
```

```python
#To view the first 5 lines of the dataset
data.head()
```

```python
#To print information of all columns
print(data.info())
```

```python
data['gender']= data.gender.astype(object)
data['cholesterol']= data.cholesterol.astype(object)
data['gluc']= data.gluc.astype(object)
data['smoke']= data.smoke.astype(object)
data['alco']= data.alco.astype(object)
data['active']= data.active.astype(object)
data['cardio']= data.cardio.astype(object)
```

```python
print(data.info())
```

```python
#To see whether we have any missing values
data.isnull().sum()
```

We can see that our dataset does not have any missing values
